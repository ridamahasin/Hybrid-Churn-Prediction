```python
# Step 1: Import Libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Step 2: Load the Dataset
df = pd.read_csv('bank.csv')
print("Shape:", df.shape)
print("First 5 rows:")
display(df.head())
```

```python
# Step 3: Check for Missing Values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Step 4: Remove Duplicates (if any)
df = df.drop_duplicates()
print("\nShape after removing duplicates:", df.shape)
```

```python
# Step 5: Handle Categorical Columns (Convert to Numbers)
# Identify object columns
object_cols = df.select_dtypes(include='object').columns
print("\nCategorical Columns:", list(object_cols))
# Use LabelEncoder for each categorical column
le = LabelEncoder()
for col in object_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# Step 6: Check Data Types After Encoding
print("\nData types after encoding:")
print(df.dtypes)
```

```python
# Step 7: Save Cleaned Data
df.to_csv('cleaned-dataset.csv', index=False)
print("\n✅ Cleaned dataset saved as 'cleaned-dataset.csv'")
```
